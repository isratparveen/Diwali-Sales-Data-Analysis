## Exploratory Data Analysis

### Gender

```python
df.columns
```

*Countplot - Show the counts of observations in each categorical bin.
Barplot - Show point estimates and confidence intervals using bars.*

```python
# Plotting a bar chart for Gender and it's count
ax = sns.countplot(x = 'Gender', data = df)

# 'set' sets the graph size(width, height)
sns.set(rc = {'figure.figsize':(7,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.groupby(['Gender'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
```

```python
# Gender vs Total Amount
sales_gen = df.groupby(['Gender'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)

sns.set(rc = {'figure.figsize':(7,5)})
sns.barplot(x = 'Gender', y = 'Amount', data = sales_gen)
```

*From above graph, we can see that most of the buyers are females and even the purchasing power of females are greater than men.*

### Age

```python
df.columns
```

```python
# Count vs Age group
ax = sns.countplot(data = df, x = 'Age Group', hue = 'Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.groupby(['Age Group'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
```

```python
# Total amount vs Age group
sales_age = df.groupby(['Age Group'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)

sns.barplot(x = 'Age Group', y = 'Amount', data = sales_age)
```

*From above graphs, we can see that most of the buyers are of age group 26-35 yrs females.*

### State

```python
# Total number of orders from top 10 states
sales_state = df.groupby(['State'], as_index = False)['Orders'].sum().sort_values(by = 'Orders', ascending = False).head(10)

sns.set(rc = {'figure.figsize':(16, 5)})
sns.barplot(x = 'State', y = 'Orders', data = sales_state)
```

```python
df.groupby(['State'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False).head(10)
```

```python
# Total amount from top 10 states
sales_state = df.groupby(['State'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False).head(10)

sns.set(rc = {'figure.figsize':(16, 5)})
sns.barplot(x = 'State', y = 'Amount', data = sales_state)
```

*From above graphs we can see that most of the orders and total sales/amount are from Uttar Praesh, Maharashtra and Karnataka respectively.*

### Marital Status

```python
# Count vs Marital status
ax = sns.countplot(x = 'Marital_Status', data = df)

sns.set(rc = {'figure.figsize':(5, 5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.groupby(['Marital_Status', 'Gender'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
```

```python
# Amount vs Marital status and Gender
sales_mar = df.groupby(['Marital_Status', 'Gender'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)

sns.set(rc = {'figure.figsize':(6, 5)})
sns.barplot(x = 'Marital_Status', y = 'Amount', hue = 'Gender', data = sales_mar)
```

*From above graphs, we can see that most of the buyers are married(women) and they have high purchasing power.*

### Occupation

```python
# Count vs Occupation
ax = sns.countplot(x = 'Occupation', data = df)

sns.set(rc = {'figure.figsize':(20, 5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.groupby(['Occupation'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
```

```python
# Amount vs Occupation
sales_occ = df.groupby(['Occupation'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)

sns.set(rc = {'figure.figsize':(20, 5)})
sns.barplot(data = sales_occ, x = 'Occupation', y = 'Amount')
```

*From above graphs, we can see that most of the buyers are working in IT, Healthcare and Aviation sector.*

### Product Category

```python
# Count vs Product Category
ax = sns.countplot(data = df, x = 'Product_Category')

sns.set(rc = {'figure.figsize':(26, 5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.groupby(['Product_Category'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False).head(10)
```

```python
# Amount vs Product Category
sales_pr_cat = df.groupby(['Product_Category'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False).head(10)

sns.set(rc = {'figure.figsize':(20, 5)})
sns.barplot(data = sales_pr_cat, x = 'Product_Category', y = 'Amount')
```

*From above graphs, we can see that most of the sold products are from Food, Clothing and Electronics category.*

### Product ID

```python
df.groupby(['Product_ID'], as_index = False)['Orders'].sum().sort_values(by = 'Orders', ascending = False).head(10)
```

```python
# Orders vs Product ID
sales_pr_id = df.groupby(['Product_ID'], as_index = False)['Orders'].sum().sort_values(by = 'Orders', ascending = False).head(10)

sns.set(rc = {'figure.figsize':(20, 5)})
sns.barplot(data = sales_pr_id, x = 'Product_ID', y = 'Orders')
```
